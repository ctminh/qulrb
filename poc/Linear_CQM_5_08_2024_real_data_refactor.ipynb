```python
import numpy as np
def calc_coefficients(constant: int) -> list[int]:
    num_coeffs = int(np.floor(np.log2(constant)))
    coeffs = [2 ** j for j in range(num_coeffs)]
    if constant - 2 ** num_coeffs >= 0:
        coeffs.append(constant - 2 ** num_coeffs + 1)
    return coeffs #[::-1]
```

```python
coefficients = calc_coefficients(num_of_tasks_per_process)
print(coefficients)
```

```python
variable_names = [
    (p_to, p_from, coef_id)
    for p_to_id, p_to in enumerate(processes)
    for p_from_id, p_from in enumerate(processes)
    for coef_id, coef in enumerate(coefficients)
    if p_from_id != p_to_id
]
variable_names
```

```python
# this works for the first part 
obj_function = 0
for p_to_id, p_to in enumerate(processes):
    relocated_terms = 0
    for p_from_id, p_from in enumerate(processes):
        for coefficient_id, coefficient in enumerate(coefficients):
            if p_to!=p_from:
                relocated_terms += coefficient * variables[variable_mapping[(p_to, p_from, coefficient_id)]] * in_df.loc[p_from, "w"]
    print(relocated_terms)
    bqm_by_process[p_to_id] = relocated_terms
```

```python
# now we want to have the remainder of the jobs to be subtracted for the same variable
for p_to_id, p_to in enumerate(processes):
    not_relocated_terms = num_of_tasks_per_process
    for p_from_id, p_from in enumerate(processes):
        for coefficient_id, coefficient in enumerate(coefficients):
            if p_to_id != p_from_id:
                not_relocated_terms -= coefficient * variables[variable_mapping[p_from, p_to, coefficient_id]] # this is dangerous, from->to
    not_relocated_terms *= in_df.loc[p_to, "w"]
    
    print(not_relocated_terms)
    bqm_by_process[p_to_id] += not_relocated_terms
```

constraints

1) suma po a, po b, po c ... musi by mniejsza lub rwna ni 208 (tych constraintw bdzie tyle co procesw)
2) aden makespan nie moe by gorszy ni wczeniej (tych constraintw bdzie tyle co procesw)
3) suma po wszystkich literkach musi by mniejsza ni jakie k (jeden constraint)

```python
#Constraint 1: don't exceed the number of tasks -- this should work for the other encoding

constraint_all_tasks_assigned = [0] * len(processes) #maybe processes?

for p_to_id, p_to in enumerate(processes):
    for p_from_id, p_from in enumerate(processes):
        for coefficient_id, coefficient in enumerate(coefficients):
            if p_to != p_from:
                constraint_all_tasks_assigned[p_to_id] += coefficient * variables[variable_mapping[p_from, p_to, coefficient_id]]

```

```python
constraint_all_tasks_assigned
```

```python
# Add constraint 1
for constr_1 in constraint_all_tasks_assigned:
    model.add_constraint(constr_1, "<=", num_of_tasks_per_process)
```

```python
print(model)
```

```python
bqm_by_process
```

```python
# Constraint 2

for _, entry in bqm_by_process.items():
    model.add_constraint(entry,"<=", max_load)


```

```python
print(model)
```

```python
constraint_all_tasks_assigned
```

```python
# Constraint 3
model.add_constraint(sum(constraint_all_tasks_assigned), "<=", k)
```

```python
print(model)
```
